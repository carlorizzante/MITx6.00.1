```python
import pylab as plt

sample = []
linear = []
quadratic = []
cubic = []
exponential = []

for i in range(30):
    sample.append(i)
    linear.append(i)
    quadratic.append(i**2)
    cubic.append(i**3)
    exponential.append(1.5**i)
```

```python
plt.plot(sample, linear)
plt.plot(sample, quadratic)
plt.plot(sample, cubic)
# plt.plot(sample, exponential)
```

```python
plt.show()
```

```python
plt.figure("lin-qua")
plt.title("Linear vs Quadratic")
plt.xlabel("sample")
plt.ylabel("function")
plt.plot(sample, linear, "b", label="linear", linewidth=0.5)
plt.plot(sample, quadratic, "r", label="quadratic", linewidth=2.0)
plt.legend(loc="upper left")

plt.figure("qua-cub")
plt.title("Quadratic vs Cubic")
plt.xlabel("sample")
plt.ylabel("function")
plt.plot(sample, quadratic, "g^", label="quadratic")
plt.plot(sample, cubic, "r--", label="cubic")
plt.legend(loc="upper left")

plt.figure("cub-exp")
plt.title("Cubic vs Exponential")
plt.xlabel("sample")
plt.ylabel("function")
plt.plot(sample, cubic, label="cubic")
plt.plot(sample, exponential, label="exponential")
plt.legend(loc="upper left")
plt.yscale("log")

plt.show()
```
