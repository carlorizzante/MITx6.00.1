- Planning for retirement
- Intend to save an amount m each month
- earn % r of income on investments each month
- calc retirement fund over time

```python
import pylab as plt
```

```python
def retire(monthly, rate, terms):
    savings = [0]
    base = [0]
    monthly_rate = rate/12
    for i in range(1, terms + 1):
        base += [i]
        savings += [savings[-1] * (1 + monthly_rate) + monthly]
    return base, savings    
```

```python
retire(100, 0.03, 12)
```

```python
def displayRetireWMonthlies(monthlies, rate, terms):
    plt.figure("retireMonth")
    plt.clf()
    for monthly in monthlies:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, label="retire: " + str(monthly) + " - " + str(int(yvals[terms])))
        plt.legend(loc="upper left")
    plt.show()
#     return
```

```python
displayRetireWMonthlies([500, 600, 700, 800, 900, 1000, 1100], 0.05, 40*12)
```

```python
def displayRetireWRates(monthly, rates, terms):
    plt.figure("retireRate")
    plt.clf()
    for rate in rates:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, label="retire: " + str(monthly) + " : " + str(int(rate*100)) + "%")
        plt.legend(loc="upper left")
    plt.show()
```

```python
displayRetireWRates(800, [0.03, 0.05, 0.07], 40*12)
```

```python
def displayRetireWMonthAndRate(monthlies, rates, terms):
    plt.figure("retireBoth")
    plt.clf()
    plt.xlim(30*12, 40*12)
    for monthly in monthlies:
        for rate in rates:
            xvals, yvals = retire(monthly, rate, terms)
            plt.plot(xvals, yvals, label="retire: " + str(monthly) + " : " + str(int(rate*100)) + "%")
            plt.legend(loc="upper left")
    plt.show()
```

```python
displayRetireWMonthAndRate([500, 700, 900], [0.03, 0.05, 0.07], 40*12)
```

```python
def displayRetireWMonthAndRateLabels(monthlies, rates, terms):
    plt.figure("retireBoth")
    plt.clf()
    plt.xlim(30*12, 40*12)
    labels = ["r", "b", "g", "y"]
    for i in range(len(monthlies)):
        label = labels[i%len(monthlies)]
        for rate in rates:
            xvals, yvals = retire(monthlies[i], rate, terms)
            plt.plot(xvals, yvals, label, label="retire: " + str(monthlies[i]) + " : " + str(int(rate*100)) + "%")
            plt.legend(loc="upper left")
    plt.show()
```

```python
displayRetireWMonthAndRateLabels([500, 700, 900], [0.03, 0.05, 0.07], 40*12)
```
